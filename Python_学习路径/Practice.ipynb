```python
# dont run too slow
def anagramSolution1(s1,s2):
    alist = list(s2)
    
    stillOK = True
    pos1 = 0
    
    while pos1 < len(s1):
        pos2 = 0
        found = False
        
        while pos2 < len(alist) and not found:
            if s1[pos1] == alist[pos2]:
                found = True
            else:
                pos2 += 1
        
        if found:
            alist[pos2] = None
        else:
            stillOK = False
        
    return stillOK
```

```python
anagramSolution1('abcd','cbda')
```

```python
def sort_and_compare(s1,s2):
    s1 = list(s1)
    s2 = list(s2)
    
    s1.sort()
    s2.sort()
    
    pos = 0
    matches = True
    
    while pos < len(s1) and matches:
        if s1[pos] != s2[pos]:
            matches = False
        else:
            pos += 1
    
    return matches
```

```python
sort_and_compare('abcd','cbda')
```

```python
def counter_method(s1,s2):
    c1 = [0] * 26
    c2 = [0] * 26
    
    for i in range(len(s1)):
        pos = ord(s1[i]) - ord('a')
        c1[pos] += 1
    
    for i in range(len(s2)):
        pos = ord(s2[i]) - ord('a')
        c2[pos] += 1
        
    j = 0
    stillok = True
    
    while j < len(c1):
        if c1[j] == c2[j]:
            j += 1
        else:
            stillok = False
    
    return stillok
```

```python
hash_map = {}
for val in nums:
    if val not in hash_map:
        hash_map[val] = 1
    else:
        hash_map[val] += 1

hash_map
```

```python
max_num = [k for k,v in hash_map.items() if v == max(hash_map.values())]
max_num
```

```python
l,r = 0, len(nums) -1
res = float('inf')
num_length = []
for val in max_num:
    found = False
    temp = 0
    while l < r and not found:
        if nums[l] == val and nums[r] == val:
            found = True
            temp = r - l + 1
        elif nums[l] != val:
            l += 1
        elif nums[r] != val:
            r -= 1
    num_length.append(temp)
    if temp < res:
        res = temp
    
res, num_length
```

```python
def perfect(num):
    if num < 0: return False
    if num == 0: return True
    
    l, r = 0, num
    
    while l < r:
        mid = (l + r) // 2
        if mid ** 2 == num:
            return True
        elif mid ** 2 < num:
            l = mid + 1
        else:
            r = mid
            
    return False
```

```python
perfect(16)
```

```python
from collections import Counter

n = int(input())
size = list(map(int, input().split()))
shoe_left = dict(Counter(size))
customer = int(input())
profit = 0
for _ in range(customer):
    size_wanted, price = list(map(int,input().split()))
    if size_wanted in shoe_left.keys() and shoe_left[size_wanted] > 0:
        shoe_left[size_wanted] -= 1
        profit += price
    
print(profit)
```
