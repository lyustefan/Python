###   Guided Project: Visualizing Earnings Based On College Majors

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

#### Read in dataset and perform a preliminary exploration on the data

```python
recent_grads = pd.read_csv("recent-grads.csv")
print(recent_grads.iloc[0])
print(recent_grads.head())
print(recent_grads.tail())
recent_grads.describe()
# There are 173 rows in the dataframe
```

```python
# Remove rows with missing values
recent_grads = recent_grads.dropna()
```

#### Scatter plots

```python
x_name = ["Sample_size","Sample_size","Full_time","ShareWomen","ShareWomen","Men","Women"]
y_name = ["Median","Unemployment_rate","Median","Unemployment_rate","Median","Median"]

for i in range(6):
    recent_grads.plot(x=x_name[i],y=y_name[i],kind="scatter",figsize=(5,10))
```

```python
recent_grads.plot(x="Total",y="Median",kind="scatter",figsize = (5,10))
```

There is no clear linear relationship between popular majors and income level.
It appears that increasing in women share leads to a decrease in median income.
There is no clear linear relationship between full time employee and median income.

#### Histogram

```python
col_name = ["Sample_size","Median","Employed","Full_time","ShareWomen","Unemployment_rate","Men","Women"]
fig = plt.figure(figsize=(10,30))
for i in range(len(col_name)):
    ax = plt.subplot(8,1,i+1)
    ax = recent_grads[col_name[i]].plot(kind="hist")
    ax.set_title(col_name[i])
```

It appears that half of the majors are predominantly men. The most common median salary range is $ 30,000 - $ 40000
