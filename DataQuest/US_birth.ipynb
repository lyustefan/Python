```python
# Create a function to read CSV files into Python
def read_csv(file_name):
    f = open(file_name,"r")
    split_data = f.read()
    split_data = split_data.split("\n")
    # Remove header row
    string_list = split_data[1:len(split_data)]
    # Create an empty list
    final_list = []
    # Create a for loop
    for each in string_list:
        int_fields = []
        string_fields = each.split(",")
        for i in string_fields:
            int_fields.append(int(i))
        final_list.append(int_fields) 
    return final_list
```

```python
cdc_list = read_csv("US_births_1994-2003_CDC_NCHS.csv")
```

```python
# Month births counter
def month_births(input_list):
    births_per_month = {}
    for each in input_list:
        month = each[1]
        births = each[4]
        if month in births_per_month:
            births_per_month[month] = births_per_month[month] + births
        else:
            births_per_month[month] = births
    return births_per_month
```

```python
cdc_month_births = month_births(cdc_list)
```

```python
print(cdc_month_births)
```

```python
# Day of week births counter
def dow_births(input_list):
    births_dow = {}
    for each in input_list:
        dow = each[3]
        births = each[4]
        if dow in births_dow:
            births_dow[dow] = births_dow[dow] + births
        else:
            births_dow[dow] = births
    return births_dow
```

```python
cdc_day_births = dow_births(cdc_list)
print(cdc_day_births)
```

```python
def calc_counts(data,column):
    birth_count = {}
    for each in data:
        a = each[column]
        births = each[4]
        if a in birth_count:
            birth_count[a] = birth_count[a] + births
        else:
            birth_count[a] = births
    return birth_count
```

```python
cdc_year_births = calc_counts(cdc_list,0)
cdc_month_births = calc_counts(cdc_list,1)
cdc_dom_births = calc_counts(cdc_list,2)
cdc_dow_births = calc_counts(cdc_list,3)
```

Write a function that calculates the min and max value for input dictionary

```python
def min_max(input_dict,if_max = True):
    if if_max == True: 
        value = 0
        for each in input_dict:
            if input_dict[each] > value:
                value = input_dict[each]
    else:
        value = float('inf')
        for each in input_dict:
            if input_dict[each] < value:
                value = input_dict[each]
    return value
```
